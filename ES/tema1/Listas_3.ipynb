# Ejercicio 1

+ Tenemos una lista que contiene elementos de varios tipos.
+ Necesitamos hacer un programa que nos divida la lista original en 3 listas, donde cada una contiene solamente los elementos de un tipo (int, float, str) ignorando los elementos del resto de tipos.
+ Cuando se encuentre un elemento que no sea int, float, str, imprimir un mensaje "Elemento no válido -> elemento"
+ Imprimir las 3 listas para comprobar el resultado.


Lista inicial ['h', 2, 3.5, 5, [9],  'j', 4, 7.8, 'o']

Resultado:

+ lista_int = [2, 5, 4]
+ lista_float = [3.5, 7.8]
+ lista_str = ['h', 'j', 'o']


# Ayuda

La función **isinstance()** en Python sirve para determinar si un objeto es una instancia de una clase específica. 

Sintaxis: isinstance(objeto, clase)

```python
lista = ['h', 2, 3.5, 5, [9], 'j', 4, 7.8, 'o']

lista_str = []
lista_int = []
lista_float = []

for i in lista:
    if isinstance(i, str):
        lista_str.append(i)
    elif isinstance(i, int):
        lista_int.append(i)
    elif isinstance(i, float):
        lista_float.append(i)
print(f"Lista 'int' {lista_int}")
print(f"Lista 'float' {lista_float}")
print(f"Lista 'str' {lista_str}")
```

# Ejercicio 2: Agenda de contactos

+ Crea una agenda de contactos desarollando varias funciones. 
+ La agenda se compone de 2 listas, una lista de nombres y una lista de números de teléfonos. Ambas listas se relacionan por el índice, es decir, si Juán ocupa el índice 5 en la lista de nombres, su número será el índice 5 de la lista de números.
+ Crear las siguientes funciones para manejar la lista:
    + **Insertar contacto**. Se debe pedir al usuario, dentro de la función, que introduzca el nombre y el número. Si el nombre ya existe, el usuario debe tener 3 opciones: actualizar el número; introduzca un nombre diferente; ignorar la operación. La agenda en todo momento debe estar orenada por nombre alfabéticamente, por lo que cada nuevo contacto se debe insertar en su posición correcta. 
    + **Mostrar agenda**.
    + **Buscar contacto**: Se debe permitir buscar por nombre o por número de teléfono. Y mostrar el otro dato, es decir, si se busca por número, mostrar el nombre a quién pertenece.
    + **Eliminar contacto**: Preguntar al usuario que nombre desea eliminar (no olvidar borrar también el número). Si no existe el contacto se debe informar de ello.
    + **Borrar toda la agenda**

```python
def introducir_nuevo(nombres, numeros, nuevo_nombre, nuevo_numero):
    # Buscamos la posición (índice) correcto según orden alfabético
    i = 0
    while i < len(nombres) and nuevo_nombre > nombres[i]:
        i += 1
    # Insertamos en la posición índice tanto el nombre como el número    
    nombres.insert(i, nuevo_nombre)
    numeros.insert(i, nuevo_numero)
    print("Contacto agregado.")


def insertar_contacto(nombres, numeros):
    nuevo_nombre = input("Ingrese el nombre: ")
    nuevo_numero = input("Ingrese el número de teléfono: ")

    # Si el nombre no existe ya en la lista, lo insertamos
    if nuevo_nombre not in nombres:
        introducir_nuevo(nombres, numeros, nuevo_nombre, nuevo_numero)
        
    # Si el nombre ya existe    
    else:
        print(f"El nombre '{nuevo_nombre}' existe en la agenda. ¿Qué desea hacer?")
        opcion = input("1. Actualizar número\n2. Introducir otro nombre\n3. Ignorar\n")
        
        # Actualizar número
        if opcion == "1":
            # Buscamos el índice del nombre que ya existe
            index = nombres.index(nuevo_nombre)
            # Actualizamos número
            numeros[index] = nuevo_numero
            print("Número actualizado.")
        
        # Introducir nuevo nombre que no exista
        elif opcion == "2":
            # Pedimos un nuevo nombre hasta que este no exista
            salir = False
            while not salir:
                nuevo_nombre = input("Introduzca un nuevo nombre")
                if nuevo_nombre not in nombres:
                    # Insertamos
                    introducir_nuevo(nombres, numeros, nuevo_nombre, nuevo_numero)
                    salir = True
                else:
                    # Este nuevo nombre también existe, pedimos otro
                    print(f"El nombre '{nuevo_nombre}' existe en la agenda")
        
        # Ignorar
        else:
            print("Operación ignorada.")   


def mostrar_agenda(nombres, numeros):
    if not nombres:
        print("La agenda está vacía.")
    else:
        print("Agenda de contactos:")
        for i in range(len(nombres)):
            print(f"{nombres[i]}: {numeros[i]}")
            

def buscar_contacto(nombres, numeros):
    
    opcion = int(input("1. Buscar por nombres\n2. Buscar por número\nSeleccione opción: "))
    
    if opcion == 1:
        nombre = input("Ingrese el nombre a buscar: ")
        if nombre in nombres:
            indice = nombres.index(nombre)
            print(f"El número de {nombre} es: {numeros[indice]}")
        else:
            print("Contacto no encontrado.")
    elif opcion == 2:
        numero = input("Ingrese el número a buscar: ")
        if numero in numeros:
            indice = numeros.index(numero)
            print(f"El número {numero} pertenece a {nombres[indice]}")
        else:
            print("Contacto no encontrado")
    else:
        print("Opción inválida.")

        
def eliminar_contacto(nombres, numeros):
    nombre = input("Ingrese el nombre a eliminar: ")
    if nombre in nombres:
        indice = nombres.index(nombre)
        del numeros[indice]
        del nombres[indice]
        print("Contacto eliminado.")
    else:
        print("Contacto no encontrado.")

        
def borrar_agenda(nombres, numeros):
    nombres.clear()
    numeros.clear()
    print("Agenda borrada.")

# Inicialización de las listas
nombres = []
numeros = []

# Ejemplo de uso:
insertar_contacto(nombres, numeros)
insertar_contacto(nombres, numeros)
mostrar_agenda(nombres, numeros)
buscar_contacto(nombres, numeros)
eliminar_contacto(nombres, numeros)
mostrar_agenda(nombres, numeros)
```
