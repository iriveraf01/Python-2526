# Ejercicio 1. Limpieza de texto

Escribe una función que tome una cadena de texto como entrada y devuelva la misma cadena pero sin signos de puntuación, en minúsculas y con todas las palabras separadas por un espacio. Utiliza los métodos replace(), lower(), split() y join().

Los signos de puntuación están en strings.punctuation

Ejemplo:
+ texto_sucio = "Hola, ¿cómo estás? ¡Muy bien!."
+ Salida: "hola cómo estás muy bien"

```python
def ej1(cadena):
    return cadena.replace("¿", "").replace("?", "").replace("¡", "").replace("!", "").replace(".", "").replace("," , "").lower()
texto_sucio = "Hola, ¿cómo estás? ¡Muy bien!."
ej1(texto_sucio)
```

# Ejercicio 2

Escribe un programa que pida al usuario una frase y cuente cuántas palabras existen de cada tamaño, es decir, cuántas palabras hay de 1 letra, cuántas de 2, cuántas de 3...Debes eliminar antes todos los símbolos de puntuación como puntos, comas, interrogaciones, etc (puedes aprovechar trabajo el ejercicio anterior).

Posteriormente muestra el resultado del análisis.

```python
frase = "¡Hola! ¿Cómo estás? Espero que estés bien. ¡Adiós!"
frase_limpia = ej1(frase)
palabras = frase_limpia.split()
contador_tamaños = {}
for palabra in palabras:
    tamaño = len(palabra)
    if tamaño in contador_tamaños:
        contador_tamaños[tamaño] += 1
    else:
        contador_tamaños[tamaño] = 1
print(contador_tamaños)
```

# Ejercicio 3

Crea una función que genere una contraseña aleatoria de al menos 8 caracteres, utilizando letras mayúsculas y minúsculas, números y símbolos almacenados en una lista predefinida.

Utiliza la función **choice** de la librería **random** para seleccionar caracteres aleatoriamente de una cadena.

https://docs.python.org/3/library/random.html#random.choice

```python
caracteres = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*()"
import random
def generar_contraseña(longitud):
    contraseña = ""
    for i in range(longitud):
        contraseña += random.choice(caracteres)
    return contraseña
print(generar_contraseña(12))
```

# Ejercicio 4

+ El ejercicio trata de simular el lanzamiento de un dado de 6 caras utilizando bucle y la función **random.randint**. 
+ El programa debe permitir al usuario elegir cuántas veces quiere lanzar el dado.
+ Posteriormente, lanzar el dado las veces que ha elegido el usuario, contando la frecuencia de aparición de cada número. Es decir, contar cuántas veces sale el 1, cuántas sale el 2...
+ Finalmente mostrar la tabla de frecuencias.

```python
import random
dic = {}
n = int(input("¿Cuántas veces quieres lanzar el dado? "))
for i in range(n):
    lanzamiento = random.randint(1, 6)
    if lanzamiento in dic:
        dic[lanzamiento] += 1
    else:
        dic[lanzamiento] = 1
print("Frecuencia de aparición de cada número:")
for numero in range(1, 7):
    frecuencia = dic.get(numero, 0)
    print(f"Número {numero}: {frecuencia} veces")
```

# Ejercicio 5

+ Crea una función llamada contar_palabras que reciba un texto como parámetro y cuente la frecuencia de cada palabra en el texto. Debes eliminar primero los signos de puntuación (están como constante en la librería string.punctuation).
+ La función a realizar debe retornar un diccionario que almacena las palabras como claves y su frecuencia como valores. Ejemplo: Para la frase “hola hola python” devolver {“hola”: 2, “python”: 1}
+ Fuera de la función, en el “main” invoca a la función anterior y muestra el resultado obtenido recorriendo el diccionario con un for.


Puede ayudar el uso de las funciones de la clase str “maketrans” y “translate”.

```python
# Texto de ejemplo
texto = "¿Este es un ejemplo de texto? para ¡contar! palabras. Las palabras pueden - repetirse."

import string
def contar_palabras(texto):
    # Eliminar signos de puntuación
    texto_limpio = texto.translate(str.maketrans("", "", string.punctuation))
    # Convertir a minúsculas y dividir en palabras
    palabras = texto_limpio.lower().split()
    # Contar frecuencia de cada palabra
    frecuencia_palabras = {}
    for palabra in palabras:
        if palabra in frecuencia_palabras:
            frecuencia_palabras[palabra] += 1
        else:
            frecuencia_palabras[palabra] = 1
    return frecuencia_palabras

resultado = contar_palabras(texto)
for palabra, frecuencia in resultado.items():
    print(f'Palabra: "{palabra}", Frecuencia: {frecuencia}')
```
