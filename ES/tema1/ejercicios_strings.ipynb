# EJERCICIO 3: VALIDACIÓN SIMPLE SIN USAR REGEX

+ Pide al usuario que introduzca un correo electrónico y verifica si contiene un formato adecuado.

+ El formato a comprobar (simplificado) será nombre@dominio.es o nombre@dominio.com:

    + El nombre puede ser cualquier cadena sin @
    + El dominio puede ser cualquier cadena sin @ ni .
    + Debe acabar obligatoriamente en .com o .es. 
    + Es decir: 
        + Solo puede haber un @ en toda la cadena. 
        + A la derecha de la @ solo puede haber 1 punto que sirve para separar el dominio.

+ A cada tipo de error mostrar un mensaje personalizado que informe al usuario
+ No es case-sensitive
+ Métodos a usar: lower, endswith, split.

Se proponen unos if para organizar el código:
+ Comprobar que solo tenga 1 @ y que acaba en .com o .es. Si no lo cumple mostrar "El formato debe ser..."
+ Comprobar que el nombre no esté vacío. Si no lo cumple mostrar "El nombre está vacío"
+ Comprobar que después del @ solo tiene 1 punto. Si tiene más mostrar el error.
+ Comprobar que dominio no esté vacío.

Si el correo es correcto mostrar el mensaje.
