```python
# 1. deque. Simulación historial navegador especial. 
 
#   Disponemos de 2 colas que almacenan un máximo de 10 entradas: 
# ●  historial: Almacena páginas visitadas para poder volver a ellas Si “pulsamos” hacia 
# atrás <-. La última página introducida en el historial es la actualmente visible.  
# ●  revertidos: Almacena páginas revertidas. Si estamos en B le damos hacia atrás (<-), 
# B se guarda en revertidos para poder volver si pulsamos ->. 
 
# Hacer un script con ambas estructuras como variables globales. Funciones a realizar: 
# ●  visitar_pagina(url): 
# ○  Añade la url en el historial y borra revertidos. Por lo tanto, la página visible 
# actual es la última en ser introducida en el historial.  
# ●  ir_atrás(): 
# ○  Si en el historial existe más de una página para poder “ir atrás”: 
# ■  Saca la actual del historial y la introduce en revertidos.  
# ■  Muestra por pantalla la nueva actual visible.  
# ○  Si no existen páginas para ir atrás se muestra mensaje. 
# ●  ir_adelante(): 
# ○  Si existe página en revertidos, saca la última en ser introducida y la mete en 
# historial.  
# ○  Si no existe página a la que ir se debe mostrar mensaje. 
 
#   Se debe mostrar mensajes para saber que pasa en cada momento. Hacer un “main” 
# con pruebas de las 3 funciones.
from collections import deque

historial = deque(maxlen=10)
revertidos = deque(maxlen=10)
def visitar_pagina(url):
    historial.append(url)
    revertidos.clear()
    print(f"Página visitada: {url}")

def ir_atrás():
    if len(historial) > 1:
        pagina_actual = historial.pop()
        revertidos.append(pagina_actual)
        print(f"Página actual después de ir atrás: {historial[-1]}")
    else:
        print("No hay páginas para ir atrás.")
        
def ir_adelante():
    if revertidos:
        pagina_revertida = revertidos.pop()
        historial.append(pagina_revertida)
        print(f"Página actual después de ir adelante: {pagina_revertida}")
    else:
        print("No hay páginas para ir adelante.")

# Pruebas
if __name__ == "__main__":
    visitar_pagina("pagina1.com")
    visitar_pagina("pagina2.com")
    visitar_pagina("pagina3.com")
    
    ir_atrás()  # Debería mostrar pagina2.com
    ir_atrás()  # Debería mostrar pagina1.com
    ir_atrás()  # No hay páginas para ir atrás.
    
    ir_adelante()  # Debería mostrar pagina2.com
    ir_adelante()  # Debería mostrar pagina3.com
    ir_adelante()  # No hay páginas para ir adelante.
    
    visitar_pagina("pagina4.com")  # Visitar nueva página, debería limpiar revertidos
    ir_atrás()  # Debería mostrar pagina3.com
```

```python
# 2. Counter. Dado un texto, cuenta las palabras (ignorando mayúsculas/minúsculas, sin 
# limpiar signos de puntuación ni tildes)  y muestra: 
# 1.  Las 5 palabras más frecuentes 
# 2.  Cuántas palabras distintas hay 
# 3.  Cuántas veces aparece una palabra Python. 
# Texto a limpiar: "Python es genial y PYTHON es práctico. Aprender python 
# es útil"

from collections import Counter
def analizar_texto(texto):
    palabras = texto.lower().split()
    contador = Counter(palabras)
    
    print("Las 5 palabras más frecuentes:")
    for palabra, frecuencia in contador.most_common(5):
        print(f"{palabra}: {frecuencia}")
    
    print(f"\nNúmero de palabras distintas: {len(contador)}")
    print(f"Número de veces que aparece 'python': {contador['python']}")
# Prueba
if __name__ == "__main__":
    texto = "Python es genial y PYTHON es práctico. Aprender python es útil"
    analizar_texto(texto)
```
