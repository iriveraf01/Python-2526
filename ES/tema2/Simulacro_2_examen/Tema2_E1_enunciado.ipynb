## Ejercicio 1: Lotería

Crea una **función llamada "premios"** para simular un sorteo de lotería de 2 cifras (del 10 al 99). Se harán 20 tiradas. Se debe utilizar el módulo **random** para generar las tiradas. Cada tirada es una tupla, todas las tiradas se almacenan en un conjunto y se retornan. No puede haber 2 números iguales, es decir, si ya ha salido el 54, no puede volver a salir... Usad __conjuntos__

Crea otra **función llamada "apuesta"** con 5 tiradas aleatorias. Estas serán los números que juega el jugador. 

Crea otra función llamada **"comprobación"** que compara ambos conjuntos para saber cuántos números ha acertado. Debes hacer la comprobación mediante operaciones de conjuntos. Debes retornar una tupla con la cantidad de números acertados y la cantidad de números no acertados, es decir (acertados, no_acertados).

```python
import random
# Crea una **función llamada "premios"** para simular un sorteo de lotería de 2 cifras (del 10 al 99). Se harán 20 tiradas. Se debe utilizar el módulo **random** para generar las tiradas. Cada tirada es una tupla, todas las tiradas se almacenan en un conjunto y se retornan. No puede haber 2 números iguales, es decir, si ya ha salido el 54, no puede volver a salir... Usad __conjuntos__

# Crea otra **función llamada "apuesta"** con 5 tiradas aleatorias. Estas serán los números que juega el jugador. 

# Crea otra función llamada **"comprobación"** que compara ambos conjuntos para saber cuántos números ha acertado. Debes hacer la comprobación mediante operaciones de conjuntos. Debes retornar una tupla con la cantidad de números acertados y la cantidad de números no acertados, es decir (acertados, no_acertados).
def premios():
    tiradas = set()
    while len(tiradas) < 20:
        tirada = random.randint(10, 99)
        tiradas.add(tirada)
    return tiradas

def apuesta():
    jugada = set()
    while len(jugada) < 5:
        numero = random.randint(10, 99)
        jugada.add(numero)
    return jugada

def comprobacion(tiradas, jugada):
    acertados = tiradas.intersection(jugada)
    no_acertados = jugada.difference(tiradas)
    return (len(acertados), len(no_acertados))

# Ejecución del programa
premios_sorteo = premios()
jugada_jugador = apuesta()
resultado = comprobacion(premios_sorteo, jugada_jugador)
print("Números premiados:", premios_sorteo)
print("Números jugados:", jugada_jugador)
print("Resultado (acertados, no acertados):", resultado)
```

## Ejercicio 2. Agenda telefónica

Implementa una agenda telefónica simple utilizando un diccionario. Las claves serán los nombres y los valores serán tuplas con el número de teléfono y la dirección.

Debes crear las siguientes funciones:

+ Introducir un nuevo contacto
+ Buscar un contacto por nombre y mostrar su teléfono y dirección.
+ Eliminar contacto por nombre
+ Mostrar toda la agenda
+ Eliminar toda la agenda

```python
# Implementa una agenda telefónica simple utilizando un diccionario. Las claves serán los nombres y los valores serán tuplas con el número de teléfono y la dirección.

# Debes crear las siguientes funciones:

# + Introducir un nuevo contacto
# + Buscar un contacto por nombre y mostrar su teléfono y dirección.
# + Eliminar contacto por nombre
# + Mostrar toda la agenda
# + Eliminar toda la agenda

agenda = {}
def introducir_contacto(nombre, telefono, direccion):
    agenda[nombre] = (telefono, direccion)
def buscar_contacto(nombre):
    return agenda.get(nombre, "Contacto no encontrado")
def eliminar_contacto(nombre):
    if nombre in agenda:
        del agenda[nombre]
def mostrar_agenda():
    return agenda.items()
def eliminar_agenda():
    agenda.clear()
# Ejecución de la agenda telefónica
introducir_contacto("Juan", "123456789", "Calle Falsa 123")
introducir_contacto("Ana", "987654321", "Avenida Siempre Viva 742")
print("Buscar contacto Juan:", buscar_contacto("Juan"))
eliminar_contacto("Ana")
print("Agenda completa:", list(mostrar_agenda()))
```

```python
# 3. namedtuple. Crea un namedtuple llamado Alumno con campos nombre, grupo, 
# nota.  A partir de una lista de alumnos como variable global, haz las siguientes funciones:  
# 1.  insertar_alumno(): Insertar un nuevo alumno recibidos por parámetros el nombre, 
# grupo y nota. 
# 1.  mayor_nota() Muestra los datos del alumno con mayor nota.  
# 2.  media_grupos(): Devuelve la media de notas por grupo en forma de dict. Usa un 
# defaultdict para construirlo. 
# Realiza un main que inserte algunos alumnos y probar las funciones. Alumnos de prueba: 
# ("Ana", "1A", 7.5),      ("Luis", "1A", 6.0),     ("Marta", "1B", 8.25),     ("Juan", "1B", 9.0) 
from collections import namedtuple, defaultdict
Alumno = namedtuple('Alumno', ['nombre', 'grupo', 'nota'])
alumnos = []
def insertar_alumno(nombre, grupo, nota):
    alumno = Alumno(nombre, grupo, nota)
    alumnos.append(alumno)
    
def mayor_nota():
    return max(alumnos, key=lambda alumno: alumno.nota)

def media_grupos():
    grupos = defaultdict(list)
    for alumno in alumnos:
        grupos[alumno.grupo].append(alumno.nota)
    medias = {grupo: sum(notas)/len(notas) for grupo, notas in grupos.items()}
    return medias

# Ejecución del programa de alumnos
insertar_alumno("Ana", "1A", 7.5)
insertar_alumno("Luis", "1A", 6.0)
insertar_alumno("Marta", "1B", 8.25)
insertar_alumno("Juan", "1B", 9.0)
print("Alumno con mayor nota:", mayor_nota())
print("Media de notas por grupo:", media_grupos())
```
