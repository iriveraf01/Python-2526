```python
# Almacena la información de cada libro y guarda en un DataFrame los datos de la web: https://books.toscrape.com/

# Lo primero sería scrapear las URL de cada categoría. Posteriormente, de cada categoría/URL descargar los libros.

# Debes guardar:

# El nombre del libro
# Categoría
# Precio
# El stock (número de unidades, está dentro de la sección de cada libro)
# El número de estrellas que tiene en forma de entero de 0 a 5.
# La imagen
# Las imágenes se descargan con request(url de la imagen) y se guardan mediante open modo wb.
# El nombre del archivo debe ser el título del libro, quitando espacios y poniendo el título en CamelCase.
# La extensión debe ser la misma que el archivo descargado ( la debéis extraer, puede que no sea siempre jpg)
# Cada imagen se guarda en el directorio media/categoría, siendo categoría cada categoría de libro.
import requests
from pathlib import Path
url = "https://books.toscrape.com/media/cache/2c/da/2cdad67c44b002e7ead0cc35693c0e8b.jpg"
response = requests.get(url)
# 1. Descargar la imagen
img_data = requests.get(url).content
# 2. Comprobamos si existe en el directorio, sino se crea
categoria = "travel"
ruta = Path("media") / categoria
if not ruta.exists():
    ruta.mkdir(parents=True)
titulo = "SharpObjects.jpg"
ruta = ruta / titulo
# 3. Guardarla en el disco duro. El archivo se abre en modo wb (write binario)
with open(ruta, "wb") as f:
    f.write(img_data)

# En el DataFrame se almacena todo, menos la propia imagen. En su lugar se almacena la ruta de la imagen. 
# Se debe convertir precio a float, stock a int
import pandas as pd
data = {
    "titulo": ["Sharp Objects"],
    "categoria": ["travel"],
    "precio": [10.99],
    "stock": [5],
    "estrellas": [4],
    "ruta_imagen": [str(ruta)]
}
df = pd.DataFrame(data)
print(df)
```
