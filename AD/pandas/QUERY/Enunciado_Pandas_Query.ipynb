Del conjunto de datos del titanic.csv realiza las siguientes consultas usando query y sin usar query, generalmente con loc bastará. 

En algunas consultas hay que usar más cosas aparte del query. 

```python
import pandas as pd

df = pd.read_csv('titanic.csv')
df.head()
```

```python
# Consultar de los hombres, las columnas survived, pclass, age
df.query('Sex == "male"')[['Survived', 'Pclass', 'Age']].head()
df.loc[df['Sex'] == 'male', ['Survived', 'Pclass', 'Age']].head()
```

```python
# Pasajeros de clase menor a 3 (2 y 1)
df.query('Pclass < 3').head()
df.loc[df['Pclass'] < 3].head()
```

```python
# Hombres supervivientes, 2 consultas con operador and
df.query('Sex == "male" and Survived == 1').head()
df.loc[(df['Sex'] == 'male') & (df['Survived'] == 1)].head()
```

```python
# Pasajeros donde Hermanos/Conyuges es mayor a padres/hijos
df.query('SibSp > Parch').head()
df.loc[df['SibSp'] > df['Parch']].head()
```

```python
# Pasajeros de Embarked S o C
df.query('Embarked == "S" or Embarked == "C"').head()
df.loc[(df['Embarked'] == 'S') | (df['Embarked'] == 'C')].head()
```

```python
# Cuantos pasarejos existen del Embarked S, hazlo a través de una variable y un f-string
embarked_value = 'S'
df.query(f'Embarked == "{embarked_value}"').shape[0]
# df.loc[df['Embarked'] == embarked_value].shape[0]
```

```python
# Sobrevivientes mayores de edad 
df.query('Survived == 1 and Age >= 18').head()
df.loc[(df['Survived'] == 1) & (df['Age'] >= 18)].head()
```

```python
# Cuantos no supervivientes menores de edad habia de clase 3
df.query('Survived == 0 and Age < 18 and Pclass == 3').shape[0]
df.loc[(df['Survived'] == 0) & (df['Age'] < 18) & (df['Pclass'] == 3)].shape[0]
```

```python
# La edad media de los supervivientes mujeres de clase 1 o de clase 2
print(f"La edad media es --> {df.query('Survived == 1 and Sex == "female" and (Pclass == 1 or Pclass == 2)')['Age'].mean()}")
print(f"La edad media es --> {df.loc[(df['Survived'] == 1) & (df['Sex'] == 'female') & ((df['Pclass'] == 1) | (df['Pclass'] == 2)), 'Age'].mean()}")
```

```python
# Mostrar supervivientes donde la edad es desconocida
df.query('Survived == 1 and Age.isnull()').head()
df.loc[(df['Survived'] == 1) & (df['Age'].isnull())].head()
```

```python
# Mostrar aquellos que no sobrevivieron y pagaron un precio entre 15 y 20
df.query('Survived == 0 and (Fare >= 15 and Fare <= 20)').head()
df.loc[(df['Survived'] == 0) & (df['Fare'] >= 15) & (df['Fare'] <= 20)].head()
```

```python
# Consultar a través de una lista los los valores donde Embarked es S y C
embarked_list = ['S', 'C']
df.query('Embarked in @embarked_list').head()
df.loc[df['Embarked'].isin(embarked_list)].head()
```

```python
# Consultar aquellos donde el número de familiares totales es mayor a 5
df.query('(SibSp + Parch) > 5').head()
df.loc[(df['SibSp'] + df['Parch']) > 5].head()
```

```python
# Consultar cuántos supervivientes viajab sin familiares
df.query('Survived == 1 and (SibSp + Parch) == 0').shape[0]
df.loc[(df['Survived'] == 1) & ((df['SibSp'] + df['Parch']) == 0)].shape[0]
```

```python
# Consultar la mínima del hombre superviviente de tercera clase
print(f"El hombre más joven tiene {df.query('Sex == "male" and Survived == 1 and Pclass == 3')['Age'].min()} años")
print(f"El hombre más joven tiene {df.loc[(df['Sex'] == 'male') & (df['Survived'] == 1) & (df['Pclass'] == 3), 'Age'].min()} años")
```

```python
# Meter en una lista la cantidad de pasajeros por tipo de Embarked
embarked_types = df['Embarked'].unique().tolist()
passenger_counts = [df.query(f'Embarked == "{embarked}"').shape[0] for embarked in embarked_types]
print(passenger_counts)
# Meter en una lista la cantidad de pasajeros por tipo de Embarked
embarked_types = df['Embarked'].unique().tolist()
passenger_counts = [df.loc[df['Embarked'] == embarked].shape[0] for embarked in embarked_types]
print(passenger_counts)
```
