```python
import pandas as pd

df = pd.read_csv('titanic.csv', index_col=0)
df.head(3)
```

# Limpieza

```python
# De la columna edad, cambia los valores nulos a la edad media según edad y Pclass.
# Es decir, si un nulo es mujer y de clase 3, su valor en edad será la media de edades de mujeres de clase 3.
df['Age'] = df.groupby(['Sex', 'Pclass'])['Age'].transform(lambda x: x.fillna(x.mean()))
```

```python
# Cambia los nulos de Cabin por el valor más repetido según su Pclass
df['Cabin'] = df.groupby('Pclass')['Cabin'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
# Elmina las columnas con nulos en Embarked
df = df.dropna(subset=['Embarked'])
```

```python
# Elimina la columna Ticket y Cabin
df = df.drop(columns=['Ticket', 'Cabin'])
```

```python
# Renombra las etiquetas de las columnas a valores en español con rename
df = df.rename(columns={
    'Survived': 'Superviviente',
    'Pclass': 'Clase',
    'Name': 'Nombre',
    'Sex': 'Sexo',
    'Age': 'Edad',
    'SibSp': 'Hermanos/Conyuge',
    'Parch': 'Padres/Hijos',
    'Fare': 'Tarifa',
    'Embarked': 'Puerto_Embarque'})
```
