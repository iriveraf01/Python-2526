# Aritmética y alineación de datos

Pandas simplifica las operaciones entre objetos con índices diferentes, alineando las operaciones según la etiqueta del índice. 

Si se trata de un DataFrame, alinea por nombre de columna y etiqueta de fila. 

Veamos un ejemplo:

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=["a", "c", "d", "e"])
s1
```

```python
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=["a", "c", "e", "f", "g"])
s2
```

```python
# La resta se realiza coincidiendo el índice, rellenando el resto de valores con NaN
s1 - s2
```

# Métodos aritméticos y evitar valores NaN

En operaciones aritméticas entre objetos indexados de manera diferente, es posible que desee completar los valores faltantes (nan) con un valor especial, como 0, cuando se encuentra una etiqueta de eje en un objeto pero no en el otro. 

Uusando el método del DataFrame o Serie para realizar la operación aritmética podemos indicar mediante el argumento 'fill_value' que sustituirá los valores NaN para realizar la operación aritmética.

Los métodos que comienzan por r tienen los operandos del reverso.

+ suma: add, radd 
+ resta: sub, rsub 	
+ división: div, rdiv 	
+ división entera: floordiv, rfloordiv 	
+ multiplicación: mul, rmul 	
+ exponente: pow, rpow 	

```python
# Restamos a la Serie s1 los valores de s2 (s1 - s2)
s1.sub(s2, fill_value=0)
```

```python
# Con rsub hacemos lo contrario, s2 - s1
s1.rsub(s2, fill_value=0)
```

# Apply de Pandas
El método apply sirve para aplicar cualquier función definida por el programador a cada elemento de la Series, devolviendo una nueva Series con el resultado. 

También se puede aplicar a todo un DataFrame por filas o por columnas (devuelve un objeto Series). Por defecto se aplica la función a cada columna.

```python
df.head()
```

```python
# Tenemos una función que recibe una Serie y devuelve el valor máximo.
def funcion1(x):
    return x.max()

# Aplicamos la función al DataFrame (por defecto axis=0). Obteniendo el máximo de cada columna. 
df.apply(funcion1)
```

```python
# Busca en el eje horizontal, sacando el máximo de cada fila
df.apply(funcion1, axis='columns')
```

# Práctica 6

Crea una función que se aplique a los valores de badroom de sample1 un redondeo el valor al entero más cercano

```python
def redondeo(x):
    return round(x)
sample1["bedrooms"].apply(redondeo)
```

La función a aplicar con apply también puede recibir parámetros

Busca en la documentación de Pandas.

Crea una función similar a la anterior, pero que si el número de baños es mayor a 4 devuelva el entero pasado por parámetro.


```python
def redondeo_bathrooms(x, entero):
    if x > 4:
        return entero
    else:
        return round(x)
sample1["bathrooms"].apply(redondeo_bathrooms, entero=999)
```

# Práctica 9
Genera una serie con el precio medio de cada ciudad. Si necesitas bucles puedes usarlos.


```python
ciudades = corte500["city"]
precios_medios = {}
for ciudad in ciudades:
    precios_ciudad = corte500[corte500["city"] == ciudad]["price"]
    precio_medio = precios_ciudad.mean()
    precios_medios[ciudad] = precio_medio
serie_precios_medios = pd.Series(precios_medios)
serie_precios_medios
```

# Práctica 10

Genera un df a partir de sample1 con las casas más baratas que la media en función de la ciudad en la que se encuentran.


```python
# Genera un df a partir de sample1 con las casas más baratas que la media en función de la ciudad en la que se encuentran.
precios_medios_ciudad = serie_precios_medios.to_dict()
def es_mas_barata_que_media(row):
    ciudad = row['city']
    precio = row['price']
    precio_medio = precios_medios_ciudad.get(ciudad, float('inf'))
    return precio < precio_medio
casas_mas_baratas = sample1[sample1.apply(es_mas_barata_que_media, axis=1)]
casas_mas_baratas
```

# MultiIndex

__set_index__ nos permite indicar qué columna serán las etiquetas de las filas (índice de filas). Pero también podemos indicar más de una columna, obteniendo indices de varios niveles.

Si seleccionamos una columna que no es única en sus valores debemos tenerlo en cuenta, ya que tendremos un índice no único como filas.

```python
# Ejemplo
import pandas as pd

data = {
    'Ciudad': ['Madrid', 'Madrid', 'Barcelona', 'Barcelona'],
    'Año': [2023, 2024, 2023, 2024],
    'Ventas': [100, 120, 90, 110],
    'Gastos': [70, 80, 60, 75]
}
df = pd.DataFrame(data)
df.head()
```

```python
# Agrupamos datos primero por Ciudad y luego por Año
df_multi = df.set_index(['Ciudad', 'Año'])
df_multi.head()
```

```python
# Indexación en orden de jerarquía
# Todos los datos de madrid 
df_multi.loc['Madrid']
```

```python
# Indexación de filas concretas usando una tupla de índices
# Madrid 2023  
df_multi.loc[('Madrid', 2023)]
```
