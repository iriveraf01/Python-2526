```python
# 1. Cargar los archivos en un dataframe read_csv.
# 2. Indicar que la columna time es de tipo datetime df['time'] = pd.to_datetime(df['time'])
# 3. Pasar la columna time a etiqueta de las filas con set_index('time')
# 4. Indexar la columna value para obtener solamente una Series.
# 5. humedad, temperatura
import pandas as pd
# 1
df_h = pd.read_csv('humedad.csv')
df_t = pd.read_csv('temperatura.csv')
# 2
df_h['time'] = pd.to_datetime(df_h['time'])
df_t['time'] = pd.to_datetime(df_t['time'])
# 3
df_h.set_index('time', inplace=True)
df_t.set_index('time', inplace=True)
# 4
series_h = df_h['value']
series_t = df_t['value']
# 5
humedad = series_h
temperatura = series_t

# # Consultas

# 1. Consultar la humedad del día de ayer, todas las filas.
# 2. Consultar la media, mínima y máxima, y std del día de ayer, humedad y temperatura.
# 3. Consultar la máxima variación de humedad en el día entre 2 horas. Calcular las medias por hora. Luego ir comparando para buscar
# la hora donde ha variado más.
# 4. Consultar la temperatura media de las 2 últimas horas, mínima y máxima.
# 5. Calcular la media por día (todos los días de los que hay datos), humedad y temperatura. Mostrar los datos tipo tabla.

# 1
ayer_h = series_h['2025-12-02']
#print(ayer_h)
ayer_t = series_t['2025-12-02']
# 2
media_h = ayer_h.mean().round(2)
print(f"La media de humedad de ayer es de {media_h}")
media_t = ayer_t.mean().round(2)
print(f"La media de temperatura de ayer es de {media_t}")
minima_h = ayer_h.min()
print(f"La mínima de ayer de humedad es de {minima_h}")
minima_t = ayer_t.min()
print(f"La mínima de ayer de temperatura es de {minima_t}")
maxima_h = ayer_h.max()
print(f"La máxima de ayer de humedad es de {maxima_h}")
maxima_t = ayer_t.max()
print(f"La máxima de ayer de temperatura es de {maxima_t}")
std_h = ayer_h.std().round(2)
print(f"La std del día de ayer de humedad es de {std_h}")
std_t = ayer_t.std().round(2)
print(f"La std del día de ayer de temperatura es de {std_t}")
```
