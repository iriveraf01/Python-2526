## Idea general ultra simple

### **Matplotlib**
Es la librería base para hacer gráficos en Python.  
Piensa en ella como el **motor** del dibujo.

Con Matplotlib tú controlas casi todo:
- tamaño del gráfico
- título
- ejes
- colores
- leyenda
- líneas, barras, puntos...

La librería más usada es:

```python
import matplotlib.pyplot as plt
```

### **Seaborn**
Seaborn está construida **encima de Matplotlib**.  
Sirve para hacer gráficos:
- más bonitos
- más rápidos de crear
- mejores para datos en `pandas.DataFrame`

La librería más usada es:

```python
import seaborn as sns
```

### Regla fácil para el examen
- Si quieres **control total**, usa **Matplotlib**
- Si quieres **hacerlo rápido y bonito**, usa **Seaborn**
- Muchas veces puedes **mezclar ambos**


# 1. Datos de ejemplo rápidos

No hace falta explicar mucho `numpy` y `pandas`, pero sí tenerlos a mano porque casi todos los ejercicios del examen irán con arrays o DataFrames.


````python

# Arrays con numpy
x = np.arange(0, 10, 1)
y = x**2

# DataFrame con pandas
df = pd.DataFrame({
    "x": np.arange(1, 11),
    "y": [3, 5, 4, 7, 6, 8, 9, 11, 10, 12],
    "categoria": ["A", "B", "A", "B", "A", "B", "A", "B", "A", "B"],
    "valor": [10, 15, 9, 18, 11, 20, 13, 17, 12, 16]
})

df.head()
````


# 2. Matplotlib desde cero: lo básico de verdad

## La idea clave
Con Matplotlib muchas veces haces esto:

1. creas el gráfico  
2. dibujas algo (`plot`, `bar`, `scatter`...)  
3. decoras (`title`, `xlabel`, `ylabel`, `legend`, `grid`)  
4. enseñas (`show`)

## Estructura típica

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Mi gráfico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.show()
```

### Traducción fácil
- `figure(...)` → crea el lienzo
- `plot(...)` → dibuja la línea
- `title(...)` → pone título
- `xlabel(...)`, `ylabel(...)` → nombres de los ejes
- `grid(True)` → cuadrícula
- `show()` → muestra el gráfico


````python

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Gráfico de líneas básico")
plt.xlabel("x")
plt.ylabel("y = x²")
plt.grid(True)
plt.show()
````


# 3. Gráfico de líneas (`plt.plot`)

## Cuándo usarlo
Cuando quieres ver:
- evolución
- tendencia
- comportamiento continuo

## Plantilla copy-paste

```python
plt.plot(x, y, marker="o", linestyle="--", color="blue", label="datos")
plt.legend()
```

## Parámetros importantes
- `marker="o"` → pone puntitos
- `linestyle="--"` → línea discontinua
- `color="red"` → color
- `label="..."` → nombre para la leyenda


````python

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker="o", linestyle="--", color="blue", label="y = x²")
plt.title("Ejemplo de plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
````


# 4. Gráfico de barras (`plt.bar`)

## Cuándo usarlo
Cuando comparas cantidades entre categorías.

Ejemplo típico:
- ventas por producto
- alumnos por clase
- notas por grupo

## Plantilla copy-paste

```python
categorias = ["A", "B", "C"]
valores = [10, 20, 15]

plt.bar(categorias, valores)
plt.show()
```


````python

categorias = ["A", "B", "C", "D"]
valores = [10, 20, 15, 12]

plt.figure(figsize=(8, 5))
plt.bar(categorias, valores)
plt.title("Gráfico de barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
````


# 5. Histograma (`plt.hist`)

## Cuándo usarlo
Para ver **cómo se distribuyen** los datos:
- si están concentrados
- si están dispersos
- si parecen normales
- si hay valores raros

## Idea fácil
El histograma **agrupa datos en intervalos** (*bins*).

## Plantilla copy-paste

```python
datos = np.random.randn(100)
plt.hist(datos, bins=10)
plt.show()
```

## Ojo con `bins`
- pocos bins → demasiado bruto
- demasiados bins → demasiado fragmentado


````python

datos = np.random.randn(1000)

plt.figure(figsize=(8, 5))
plt.hist(datos, bins=20, edgecolor="black")
plt.title("Histograma")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
````


# 6. Diagrama de dispersión (`plt.scatter`)

## Cuándo usarlo
Para ver relación entre dos variables numéricas.

Ejemplo:
- altura vs peso
- horas de estudio vs nota
- edad vs salario

## Qué buscas
- si sube una cuando sube la otra
- si hay relación
- si hay outliers


````python

x_scatter = np.random.rand(50)
y_scatter = 2 * x_scatter + np.random.rand(50) * 0.3

plt.figure(figsize=(8, 5))
plt.scatter(x_scatter, y_scatter)
plt.title("Scatter plot")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
````


# 7. Subplots: varios gráficos en una figura

## Cuándo sale en examen
Muchísimo. Te pueden pedir comparar dos gráficos en la misma ventana.

## Forma típica

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
```

Eso significa:
- `1` fila
- `2` columnas

Luego:
- `ax[0]` → primer gráfico
- `ax[1]` → segundo gráfico


````python

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].plot(x, y, color="blue")
ax[0].set_title("Líneas")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

ax[1].bar(categorias, valores, color="orange")
ax[1].set_title("Barras")
ax[1].set_xlabel("Categoría")
ax[1].set_ylabel("Valor")

plt.tight_layout()
plt.show()
````


# 8. El estilo orientado a objetos (`fig`, `ax`)

Esto cae bastante porque es la forma **más correcta** de trabajar con Matplotlib.

## Diferencia rápida

### Forma rápida:
```python
plt.plot(x, y)
```

### Forma recomendada:
```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

## Ventaja
Tienes más control, sobre todo si hay varios gráficos.


````python

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, marker="o")
ax.set_title("Usando fig y ax")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True)

plt.show()
````


# 13. Barras con Seaborn (`sns.barplot`)

## Importante
`barplot` normalmente muestra una **estimación** (por defecto, la media) para cada categoría.

Eso significa que si tienes varias filas por categoría, te dibuja la media.

## Si solo quieres contar elementos por categoría:
usa `countplot`


````python

plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="categoria", y="valor")
plt.title("Barplot con Seaborn")
plt.show()
````


# 14. Conteo por categorías (`sns.countplot`)

## Cuándo usarlo
Cuando quieres saber cuántos hay de cada clase.

Ejemplo:
- cuántos hombres y mujeres
- cuántos aprobados y suspensos
- cuántos A y B


````python

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="categoria")
plt.title("Countplot")
plt.show()
````


# 15. Histograma con Seaborn (`sns.histplot`)


````python

plt.figure(figsize=(8, 5))
sns.histplot(df["valor"], bins=8, kde=True)
plt.title("Histplot con curva KDE")
plt.show()
````


## ¿Qué es `kde=True`?
Añade una curva suave encima del histograma para ver mejor la distribución.

No hace falta que te rayes:
- histograma = barras de frecuencia
- KDE = curva suavizada


## `hue` explicado fácil

Muy útil cuando quieres separar grupos.



# 17. Boxplot (`sns.boxplot`)

Para ver:


# 19. Heatmap (`sns.heatmap`)

## Muy importante para correlaciones
Te puede salir algo como:
> "Representa la matriz de correlación"

Entonces haces:
1. calculas correlación con pandas
2. la dibujas con `heatmap`

## Plantilla copy-paste

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
```

### Parámetros típicos
- `annot=True` → escribe los números dentro
- `cmap="coolwarm"` → mapa de color


````python

corr = df.corr(numeric_only=True)

plt.figure(figsize=(8, 5))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Mapa de calor de correlaciones")
plt.show()
````

# 23. Errores típicos de examen

## 1) Olvidar `plt.show()`
A veces en notebook se ve igual, pero en examen mejor ponerlo.

## 2) Usar nombres de columnas mal escritos
Si haces:
```python
sns.lineplot(data=df, x="X", y="y")
```
y la columna es `"x"`, fallará.

## 3) Pasar texto donde va una variable numérica
Por ejemplo, scatter necesita variables numéricas en ejes.

## 4) Confundir `barplot` con `countplot`
- `barplot` → resume una variable numérica por categoría
- `countplot` → cuenta cuántas filas hay en cada categoría

## 5) No importar librerías
Pon siempre arriba:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 6) No usar `numeric_only=True` en correlaciones
A veces el DataFrame tiene texto y da guerra.

```python
df.corr(numeric_only=True)
```

## 7) No ajustar el tamaño
Si todo queda apelotonado:

```python
plt.figure(figsize=(10, 6))
```

o en subplots:

```python
plt.tight_layout()
```

# 26. Mini ejercicio tipo examen

## Enunciado inventado
Dado un DataFrame con notas de alumnos:
1. muestra un histograma de las notas  
2. muestra un boxplot por grupo  
3. representa la correlación entre variables numéricas  

## Solución rápida


````python

df_exam = pd.DataFrame({
    "nota": [5, 7, 8, 6, 9, 4, 7, 8, 6, 10],
    "horas_estudio": [2, 3, 5, 2, 6, 1, 4, 5, 3, 7],
    "edad": [18, 19, 18, 20, 19, 18, 21, 20, 19, 22],
    "grupo": ["A", "A", "B", "A", "B", "A", "B", "B", "A", "B"]
})

# 1) Histograma de notas
plt.figure(figsize=(8, 5))
sns.histplot(df_exam["nota"], bins=6, kde=True)
plt.title("Distribución de notas")
plt.show()

# 2) Boxplot por grupo
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_exam, x="grupo", y="nota")
plt.title("Notas por grupo")
plt.show()

# 3) Correlación
corr_exam = df_exam.corr(numeric_only=True)

plt.figure(figsize=(6, 4))
sns.heatmap(corr_exam, annot=True, cmap="coolwarm")
plt.title("Correlación")
plt.show()
````
