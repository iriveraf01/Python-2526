
# 8. Matplotlib: explicación fácil otra vez

## Qué hace
Matplotlib es la librería base para gráficos.

La idea típica es:

```python
plt.figure(figsize=(8, 5))
plt.plot(...)
plt.title(...)
plt.xlabel(...)
plt.ylabel(...)
plt.show()
```

## `figsize=(8,5)` explicado fácil
Es el tamaño del gráfico:
- 8 → ancho
- 5 → alto

No hay un tamaño “mágico”.  
Se pone uno que se vea bien.



# 9. Gráfico de barras con datos reales  
## Top 10 países con más medallas totales


````python

top10_paises = df.groupby("Country_Name")["Total"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10_paises.index, top10_paises.values)
plt.title("Top 10 países con más medallas")
plt.xlabel("País")
plt.ylabel("Total de medallas")
plt.xticks(rotation=45)
plt.show()
````


## Cuándo usar barras
Cuando comparas cantidades entre categorías.  
Aquí:
- categoría = país
- valor = total de medallas



# 10. Gráfico de líneas  
## Evolución del total de medallas por año


````python

medallas_por_anyo = df.groupby("Year")["Total"].sum()

plt.figure(figsize=(10, 6))
plt.plot(medallas_por_anyo.index, medallas_por_anyo.values, marker="o")
plt.title("Evolución del total de medallas por edición")
plt.xlabel("Año")
plt.ylabel("Total de medallas")
plt.show()
````


## Cuándo usar líneas
Cuando quieres ver evolución o tendencia en el tiempo.

Aquí el eje X es el año, así que tiene sentido.

# 13. Subplots  
## Varios gráficos en la misma figura


````python

fig, ax = plt.subplots(1, 2, figsize=(14, 5))

ax[0].hist(df["Gold"], bins=15, edgecolor="black")
ax[0].set_title("Histograma de oros")
ax[0].set_xlabel("Gold")
ax[0].set_ylabel("Frecuencia")

ax[1].scatter(df["Gold"], df["Total"])
ax[1].set_title("Gold vs Total")
ax[1].set_xlabel("Gold")
ax[1].set_ylabel("Total")

plt.tight_layout()
plt.show()
````

# 15. Barplot con Seaborn  
## Top 10 países con más medallas


````python

top10_df = (
    df.groupby("Country_Name", as_index=False)["Total"]
      .sum()
      .sort_values("Total", ascending=False)
      .head(10)
)

plt.figure(figsize=(10, 6))
sns.barplot(data=top10_df, x="Country_Name", y="Total")
plt.title("Top 10 países con más medallas (Seaborn)")
plt.xlabel("País")
plt.ylabel("Total")
plt.xticks(rotation=45)
plt.show()
````


# 16. Histplot con Seaborn  
## Distribución de medallas totales


````python

plt.figure(figsize=(8, 5))
sns.histplot(df["Total"], bins=20, kde=True)
plt.title("Distribución del total de medallas")
plt.xlabel("Total")
plt.ylabel("Frecuencia")
plt.show()
````


## `kde=True` explicado fácil
Añade una curva suave para ayudarte a ver la forma de la distribución.



# 17. Scatterplot con Seaborn  
## Oro frente a plata, coloreando por país anfitrión


````python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Gold", y="Silver", hue="Host_country")
plt.title("Gold vs Silver por país anfitrión")
plt.show()
````


## `hue` explicado fácil
Sirve para colorear según una variable categórica.

# 18. Boxplot con Seaborn  
## Comparar total de medallas en un grupo de países top


````python

paises_top = top10_df["Country_Name"].tolist()
df_top = df[df["Country_Name"].isin(paises_top)]

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_top, x="Country_Name", y="Total")
plt.title("Distribución del total de medallas en países top")
plt.xlabel("País")
plt.ylabel("Total")
plt.xticks(rotation=45)
plt.show()
````


## Qué enseña un boxplot
- mediana
- dispersión
- valores raros / outliers

Muy útil cuando comparas grupos.



# 19. Heatmap de correlaciones  
## Muy típico de examen


````python

corr = df[["Year", "Gold", "Silver", "Bronze", "Total"]].corr(numeric_only=True)

plt.figure(figsize=(8, 5))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Mapa de calor de correlaciones")
plt.show()
````


## Qué significa
Mide cómo se relacionan las variables numéricas entre sí.

Ejemplo:
- si `Gold` y `Total` tienen correlación alta, tiene sentido
- porque cuantos más oros, normalmente más medallas totales

# 23. Errores típicos de examen

## 1) No crear la columna `Total`
Si te piden total de medallas, no existe de serie en este CSV.  
La tienes que crear:

```python
df["Total"] = df["Gold"] + df["Silver"] + df["Bronze"]
```

## 2) Olvidar `plt.show()`
Mejor ponerlo siempre.

## 3) Usar mal el nombre de una columna
Tiene que coincidir EXACTAMENTE.

Por ejemplo:
- `Country_Name`
- `Host_country`

No es lo mismo que:
- `country_name`
- `host_country`

## 4) No rotar etiquetas en barras
Con nombres largos de países se pisan mucho.

```python
plt.xticks(rotation=45)
```

## 5) Confundir `barplot` con `countplot`
Aquí normalmente interesa `barplot`, porque ya tienes valores numéricos.
`countplot` cuenta filas.



# 24. Mini simulación tipo examen

## Enunciado ejemplo
Con el archivo `medallas_jjoo.csv`:

1. Lee el archivo  
2. Crea una columna con el total de medallas  
3. Muestra los 10 países con más medallas  
4. Representa esa información con un gráfico de barras  
5. Muestra la matriz de correlación con un heatmap  

## Solución rápida


````python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("medallas_jjoo.csv")
df["Total"] = df["Gold"] + df["Silver"] + df["Bronze"]

top10 = (
    df.groupby("Country_Name")["Total"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10, 6))
plt.bar(top10.index, top10.values)
plt.title("Top 10 países con más medallas")
plt.xlabel("País")
plt.ylabel("Total")
plt.xticks(rotation=45)
plt.show()

corr = df[["Year", "Gold", "Silver", "Bronze", "Total"]].corr(numeric_only=True)

plt.figure(figsize=(8, 5))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlaciones")
plt.show()
````
