## üìÇ Dataset

Se proporciona el fichero **`examen_pandas_T2.csv`**, que contiene informaci√≥n sobre medallas ol√≠mpicas por pa√≠s y a√±o.

**Columnas:**
- `pais`: nombre del pa√≠s  
- `anio`: a√±o de los Juegos Ol√≠mpicos  
- `oro`: n√∫mero de medallas de oro  
- `plata`: n√∫mero de medallas de plata  
- `bronce`: n√∫mero de medallas de bronce  
- `poblacion_mill`: poblaci√≥n del pa√≠s (en millones)


## Ejercicio 1 ‚Äì Carga y exploraci√≥n (1 punto)

1. Carga el fichero CSV en un DataFrame llamado `df`.
2. Muestra las 5 primeras filas del DataFrame.
3. Muestra la informaci√≥n general del DataFrame.
4. Comprueba si existen valores nulos en el dataset.


```python
# Tu c√≥digo aqu√≠
import pandas as pd
import numpy as np
df  = pd.read_csv('examen_pandas_T2.csv')
print(df.head())
print('--------------------------------')
print(df.info())
print('--------------------------------')
df.isnull()
```

## Ejercicio 2 ‚Äì Consultas con `query` (2 puntos)

1. Muestra los pa√≠ses que en el a√±o **2020** consiguieron **m√°s de 9 medallas de oro**.
2. Muestra las filas correspondientes a **Espa√±a y Francia** utilizando una lista y el m√©todo `query`.
3. Muestra los registros donde el **n√∫mero total de medallas** (oro + plata + bronce) sea mayor de 30.


```python
# Tu c√≥digo aqu√≠
# Primero, filtrar las filas donde el a√±o es 2020
print("SIN QUERY")
medallas_2020_mas_de_9 = df["anio"] == 2020
# Segundo hago que la variable sea df con las filas filtradas
medallas_2020_mas_de_9 = df[medallas_2020_mas_de_9]
# Tercero, saco las filas donde las medallas de oro son mayores a 9
medallas_2020_mas_de_9 = medallas_2020_mas_de_9["oro"] > 9
# Cuarto, filtro el dataframe original con ambas condiciones
medallas_2020_mas_de_9 = df[medallas_2020_mas_de_9 & (df["anio"] == 2020)]
print(medallas_2020_mas_de_9)
print('--------------------------------')
esp_francia = df["pais"].isin(["Espa√±a", "Francia"])
esp_francia = df[esp_francia]
print(esp_francia)
print('--------------------------------')
medallas_mas_de_30 = (df["oro"] + df["plata"] + df["bronce"]) > 30
medallas_mas_de_30 = df[medallas_mas_de_30]
print(medallas_mas_de_30)
print('--------------------------------')
print("CON QUERY")
medallas_2020_mas_de_9_query = df.query('anio == 2020 and oro > 9')
print(medallas_2020_mas_de_9_query)
print('--------------------------------')
# Tu c√≥digo aqu√≠
# Segundo, filtro las filas donde el pa√≠s sea Espa√±a o Francia
esp_francia = df.query('pais == "Espa√±a" or pais == "Francia"')
print(esp_francia)
print('--------------------------------')
# Tu c√≥digo aqu√≠
# Tercero, filtro las filas donde la suma de las medallas sea mayor a 30
medallas_mas_de_30 = df.query('oro + plata + bronce > 30')
print(medallas_mas_de_30)
```

## Ejercicio 3 ‚Äì Transformaci√≥n de datos (2 puntos)

1. Crea una nueva columna llamada `total_medallas` que sea la suma de oro, plata y bronce.
2. Crea una columna `medallas_por_millon` dividiendo el total de medallas entre la poblaci√≥n.
3. Redondea la columna `medallas_por_millon` a 2 decimales.


```python
# Tu c√≥digo aqu√≠
df['total_medallas'] = df['oro'] + df['plata'] + df['bronce']
df['medallas_por_millon'] = df['total_medallas'] / df['poblacion_mill']
df['medallas_por_millon'] = df['medallas_por_millon'].round(2)
df
```

## Ejercicio 4 ‚Äì Discretizaci√≥n (2 puntos)

1. Discretiza la columna `total_medallas` en los siguientes grupos:
- **Bajo**: menos de 25
- **Medio**: entre 25 y 40
- **Alto**: m√°s de 40

2. Guarda el resultado en una columna llamada `nivel_medallas`.
3. Muestra cu√°ntos pa√≠ses hay en cada categor√≠a.


```python
# Tu c√≥digo aqu√≠
df["nivel_medallas"] = pd.cut(df['total_medallas'], bins=[0, 25, 40, 200], labels=["Bajo", "Medio", "Alto"])
df
```

## Ejercicio 5 ‚Äì Variables dummy (1 punto)

1. Convierte la columna `nivel_medallas` en variables dummy.
2. A√±ade estas columnas al DataFrame original usando el m√©todo `join`.


```python
# Tu c√≥digo aqu√≠
pd.get_dummies(df['nivel_medallas'], prefix='Nivel')
df = df.join(pd.get_dummies(df['nivel_medallas'], prefix='Nivel'))
df
```

## Ejercicio 6 ‚Äì Agrupaciones y agregaci√≥n (2 puntos)

1. Calcula el **total de medallas por pa√≠s**.
2. Calcula la **media de medallas de oro por pa√≠s**.
3. Calcula, agrupando por pa√≠s, el **m√°ximo y m√≠nimo** del total de medallas.


```python
# Tu c√≥digo aqu√≠
total_medallas_por_pais = df.groupby('pais')['total_medallas'].sum().reset_index()
total_medallas_por_pais = total_medallas_por_pais.sort_values(by='total_medallas', ascending=False)
print(total_medallas_por_pais)
print('--------------------------------')
media_medallas_por_pais = df.groupby('pais')['total_medallas'].mean().reset_index()
media_medallas_por_pais = media_medallas_por_pais.sort_values(by='total_medallas', ascending=False)
print(media_medallas_por_pais)
print('--------------------------------')
max_min_medallas_por_pais = df.groupby('pais')['total_medallas'].agg(['max', 'min']).reset_index()
print(max_min_medallas_por_pais)
```

## Ejercicio 7 ‚Äì Pregunta final (opcional / subida de nota)

¬øCu√°l es el pa√≠s con **mejor ratio de medallas por mill√≥n de habitantes** en todo el dataset?

Justifica brevemente el resultado mostrando los c√°lculos realizados.


```python
# Tu c√≥digo aqu√≠
# ¬øCu√°l es el pa√≠s con **mejor ratio de medallas por mill√≥n de habitantes** en todo el dataset?
mejor_ratio = df.loc[df['medallas_por_millon'].idxmax()]
print("Pa√≠s con mejor ratio de medallas por mill√≥n de habitantes:")
print(mejor_ratio)
```
