```python
import numpy as np

A = np.array([[3, 4],
              [2, 7]])

B = np.array([2, 3])

resultado = B @ np.linalg.inv(A) @ B.T

print("Resultado:", resultado)
print("Redondeado a 2 decimales:", round(resultado, 2))
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression

# Cargar dataset
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target

# Entrenar modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Obtener w0 (intercepto)
w0 = modelo.intercept_

print("w0 (intercepto):", w0)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression

# Cargar datos
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target

# Entrenar modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Coeficientes
coeficientes = modelo.coef_

print("Coeficientes:", coeficientes)

# w4 (cuarta variable)
w4 = coeficientes[3]
print("w4:", w4)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
import numpy as np

# Cargar datos
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target

# Entrenar modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Nuevo individuo
x_nuevo = np.array([
    -0.09269548, -0.04464164, -0.04069594, -0.01944183,
    -0.06899065, -0.07928784,  0.04127682, -0.0763945,
    -0.04117617, -0.09634616
])

# Predicción
pred = modelo.predict([x_nuevo])

print("Predicción:", pred[0])
```
