### Diferencias entre `.loc` y `.iloc` en Pandas

`.loc` y `.iloc` sirven para seleccionar filas y columnas, pero funcionan de forma distinta:

1. **Tipo de acceso**  
   - `.loc` usa **etiquetas** (nombres) del índice y de las columnas.  
   - `.iloc` usa **posiciones numéricas** (0, 1, 2…) como en una lista.

2. **Cómo seleccionan filas**  
   - `.loc[2]` selecciona la fila cuyo **índice se llame 2**, no necesariamente la segunda.  
   - `.iloc[2]` selecciona siempre la **tercera fila**, porque usa la posición.

3. **Rangos (slicing)**  
   - `.loc[1:3]` incluye las filas **1, 2 y 3** (incluye el final).  
   - `.iloc[1:3]` incluye las filas **1 y 2** (NO incluye el final).

4. **Índices de texto o personalizados**  
   - `.loc["Partido1"]` funciona si el índice tiene nombres.  
   - `.iloc["Partido1"]` da **error**, porque solo acepta números.

5. **Selección de columnas**  
   - `.loc[:, "Rival"]` usa el **nombre** de la columna.  
   - `.iloc[:, 2]` usa la **posición** de la columna (columna 2).

6. **Filtrado booleano**  
   - Ambos aceptan booleanos (`df.loc[df["Goles"] > 2]` y `df.iloc[df["Goles"] > 2]`),  
     pero suele usarse `.loc` porque el filtrado está basado en etiquetas.

### Ejemplos comparativos

```python
# Misma selección, diferentes sistemas:
df.loc[0:2, "Rival"]   # Etiquetas → filas 0,1,2 y columna 'Rival'
df.iloc[0:3, 2]        # Posiciones → filas 0,1,2 y columna 2


````python
# Dataframe: indexación por columnas y filas
# Selección de columnas
print("\nSelección de columna 'Rival':")
print(df["Rival"].head())
print("\nSelección de múltiples columnas 'Rival' y 'Goles_Barca':")
print(df[["Rival", "Goles_Barca"]].head())
# Selección de filas por posición
# Usando loc (índices de filas y nombres de columnas)
print("\nAcceso filas por nombre de índice con loc (filas 1 a 3, columnas 'Rival' y 'Goles_Barca'):")
print(df.loc[1:3, ["Rival", "Goles_Barca"]])
print("\nAcceso a filas por posición:")
print(df.iloc[1:4, 2:4])
````

````python
# Indexación booleana
print("\nFilas donde el Barça anotó más de 2 goles:")
print(df[df["Goles_Barca"] > 2][["Rival", "Goles_Barca", "Goles_Rival"]])
print("\nFilas donde el Barça anotó más de 2 goles y el rival más de 1:")
print(df[(df["Goles_Barca"] > 2) & (df["Goles_Rival"] == 1)][["Competición", "Rival", "Goles_Barca", "Goles_Rival"]])
````

````python
# Métodos aritméticos y evitar valores NaN
# Los métodos son: add, sub, div, floordiv (división entera), mul, pow (potencia). 
# Estos métodos permiten especificar un valor fill_value para reemplazar los valores NaN antes de realizar la operación.
# Por ejemplo, si tenemos dos DataFrames df1 y df2 y queremos sumarlos, pero df2 tiene algunos valores NaN, podemos usar fill_value para tratarlos:
df1 = pd.DataFrame({"A": [1, 2, np.nan], "B": [4, np.nan, 6]})
df2 = pd.DataFrame({"A": [np.nan, 2, 3], "B": [4, 5, np.nan]})
df_sum = df1.add(df2, fill_value=0)
print("\nSuma de df1 y df2 con fill_value=0 para NaN:")
print(df_sum)
df_sub = df1.sub(df2, fill_value=0)
print("\nResta de df1 y df2 con fill_value=0 para NaN:")
print(df_sub)
df_mul = df1.mul(df2, fill_value=1)
print("\nMultiplicación de df1 y df2 con fill_value=1 para NaN:")
print(df_mul)
df_div = df1.div(df2, fill_value=1)
print("\nDivisión de df1 y df2 con fill_value=1 para NaN:")
print(df_div)
df_pow = df1.pow(df2, fill_value=1)
print("\nPotencia de df1 y df2 con fill_value=1 para NaN:")
print(df_pow)
df_floordiv = df1.floordiv(df2, fill_value=1)
print("\nDivisión entera de df1 y df2 con fill_value=1 para NaN:")
print(df_floordiv)
````

````python
# Apply
# Aplicar una función a lo largo de un eje del DataFrame (filas o columnas)
# Por ejemplo, podemos aplicar una función que calcule el doble de cada valor en una columna
def doble(x):
    return x * 2
df_doble_goles = df["Goles_Barca"].apply(doble)
print("\nColumna 'Goles_Barca' con valores doblados usando apply:")
print(df_doble_goles)
````

````python
# # Multi index
# DataFrame.set_index() nos permitía indicar una columna para que fuese la etiqueta de las filas, “setear
# un índice”.
# Pero es posible indicar más de una columna, obteniendo índices de varios niveles. Si seleccionamos una
# columna que no es única en sus valores debemos tenerlo en cuenta, ya que tendremos un índice no
# único como filas. Lo normal es ordenar el DataFrame por el índice.
df_multi_index = df.set_index(["Competición", "Rival"])
df_multi_index = df_multi_index.sort_index()
print("\nDataFrame con multi-índice (Competición, Rival) ordenado:")
print(df_multi_index.head())
df_multi = df.set_index(["Competición", "Lugar"]).sort_index()
print("\nDataFrame con multi-índice (Competición, Lugar) ordenado:")
df_multi.head()
````
