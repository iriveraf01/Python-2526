```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.arange(5)
y = [1, 3, 2, 5, 4]

# Creamos la cuadrícula (2 filas y 2 columnas)
fig, ax = plt.subplots(2, 2, figsize=(12, 8), dpi=100)

ax[0, 0].plot(x, y)          # Línea
ax[0, 1].scatter(x, y)       # Dispersión
ax[1, 0].bar(x, y)           # Barras
ax[1, 1].hist(y)             # Histograma

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.linspace(0, 10, 100)
y = np.sin(x)
ruido = np.sin(x) + np.random.normal(0, 0.2, 100)

categorias = ['A', 'B', 'C', 'D']
valores = [3, 7, 2, 5]

normal = np.random.normal(0, 1, 1000)

# Crear subplots
fig, ax = plt.subplots(2, 2, figsize=(12, 8), dpi=100)

# 1️⃣ Seno
ax[0,0].plot(x, y, color='blue')
ax[0,0].set_title("Seno de X")
ax[0,0].grid(True)

# 2️⃣ Coseno con ruido (scatter)
ax[0,1].scatter(x, ruido, color='red', s=10)
ax[0,1].set_title("Coseno con Ruido")

# 3️⃣ Barras
ax[1,0].bar(categorias, valores, color='green')
ax[1,0].set_title("Frecuencias por Categoría")

# 4️⃣ Histograma
ax[1,1].hist(normal, bins=30, color='orange', edgecolor='black')
ax[1,1].set_title("Distribución Normal")

# Título general
fig.suptitle("Panel de Control de Análisis Exploratorio", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [100,120,90,140,160]

fig, ax = plt.subplots()

ax.plot(x, y, marker='o', linestyle='-', color='blue', label='Ventas 2024')

ax.set_title("Ventas por Mes")
ax.set_xlabel("Mes")
ax.set_ylabel("Ventas")

ax.legend()

plt.show()
```
