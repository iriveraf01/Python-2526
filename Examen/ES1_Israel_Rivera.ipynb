# Ejercicio 2 (5 puntos)

A partir de la frase inicial almacenda en la variable 'frase'

1. Convierta la frase a minúsculas y reemplace todas las vocales de la frase por el carácter 'a', imprima el resultado.
2. Divida la frase resultante en una lista de palabras. Elimine de la lista las palabras que solo sean la letra 'a' e imprima la lista.
3. De la lista anterior, crea una nueva lista con las palabras que tienen más de 4 caracteres y comienzan por consonante.
4. Muestra la lista final y genera y muestra otra lista con el número de letras que contiene cada palabra.

**Ejemplo**

frase = "Esta frase o es una frase interesante o quizás no"

Salida: 

+ Frase transformada: asta frasa a as ana frasa antarasanta a qaazás na
+ Lista palabras inicial ['asta', 'frasa', 'as', 'ana', 'frasa', 'antarasanta', 'qaazás', 'na']
+ Lista de palabras con más de 4 caracteres: ['frasa', 'frasa', 'qaazás']
+ Número de letras por palabra: [5, 5, 6]


```python
frase = "Esta frase o es una frase interesante o quizás no"

# Paso 1
frase_a = ""
for letra in frase.lower():
    if letra in "eiou":
        letra = "a"
    frase_a += letra
print(f"Frase transformada: {frase_a}")
# Paso 2
palabras_inicial = frase_a.split(" ")
for i in palabras_inicial:
    if i == "a":
        palabras_inicial.remove(i)
print(f"Lista de palabras inicial {palabras_inicial}")
# Paso 3
cuatro = []
for i in palabras_inicial:
    if i[0] not in "aeiou" and len(i) >= 4:
        cuatro.append(i)
print(f"Lista de palabras con más de 4 caracteres: {cuatro}")
# Paso 4
numero_letras = []
for i in cuatro:
    numero_letras.append(len(i))
print(f"Número de letras por palabra: {numero_letras}")
```
