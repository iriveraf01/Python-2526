# Ejercicio 1: NumPy (6 puntos)

Cada celda es 0.5 puntos a excepción de la última que vale 1 punto.

La siguiente celda solamente contiene los __datos__: 3 ndarrays con notas de alumnos y un ndarray adicional con los nombres.

```python
import numpy as np
# ======
# DATOS
#=======
nota_mates = np.array([4.3, 9.1, 7.2 , 4.8, 7.2, 4.4, 5.5, 4.2, 6.9, 3.3, 4.5, 5.6, 5.4, 6.8, 5.1])
nota_lengua = np.array([7.1, 8.7, 6.5 , 7.5, 5.6, 6.2, 5.4, 4.8 , 3.6, 4.8, 6.4, 5.6, 5.1 , 8.4, 5.0]) 
nota_fisica = np.array([7.4, 10, 7.8 , 5.4, 8.4, 3.2, 5.8, 6.4, 7.3, 3.5, 3.4, 5.6, 6.2, 4.1, 5.2])
alumnos = np.array(["Juan", "Noelia", "Pedro", "Marta", "Antonio", "Carlos", "Patricia", "Carmen", "Manuel", "Jorge", "Jose", "Sara", "María", "Rubén", "Bárbara"])
```

```python
# Crea un ndarray del mismo tamaño y tipo de datos que notas_mates llamado "nuevo" con los valores a 1
nuevo = np.ones_like(nota_mates)
print("Array de unos:\n", nuevo)

# Suma notas_mates y "nuevo" de forma vectorizada y guarda el resultado en ndarray "suma". Imprime el resultado.
suma = nota_mates + nuevo
print("\nSuma de nuevo y nota_mates")
print(suma)

# Muestra el número de notas de suma
print("\nNúmero de notas de mates: ", nota_mates.shape[0])
```

```python
# Muestra si existe al menos un alumno que haya suspendido lengua. Con mostrar True o False es sufiente.
suspenso = np.any(nota_lengua < 5)
print("Ha suspendido algún alumno: ", suspenso)
```

```python
# Muestra el nombre de los alumnos que han aprobado matemáticas
print("Alumnos aprobados en matemáticas:")
print(alumnos[nota_mates >= 5])
```

```python
# Muestra el nombre de los alumnos que han sacado, en matemáticas, menor nota que la media de matemáticas
print("Alumnos con menor nota que la media de matemáticas:")
media = np.mean(nota_mates)
print(alumnos[nota_mates < media])
```

```python
#Muestra el nombre de los alumnos que han sacado mayor nota en mates que en física
print("Alumnos que tienen mejor nota en mates que en física:")
print(alumnos[nota_mates > nota_fisica])
```

```python
# Muestra el nombre de los alumnos que han suspendido matemáticas y lengua
print("Alumnos que han suspendido matemáticas y lengua:")
print(alumnos[(nota_mates < 5) & (nota_lengua < 5)])
```

```python
# Muestra el nombre de los alumnos que han aprobado todo
print("Alumnos que han aprobado todo:")
print(alumnos[(nota_mates >= 5) & (nota_lengua >= 5) & (nota_fisica >= 5)])
```

```python
""" 
Genera un ndarray de strings, indicando en cada valor "lengua" o "fisica"
El ndarray debe indicar, por cada alumno, la asignatura donde tiene la nota más alta entre lengua y fisica
"""
# strings = np.array({"Lengua" : alumnos[nota_lengua > nota_fisica], "Física" : alumnos[nota_lengua < nota_fisica]})
# print(strings)
print(f"Lengua: {alumnos[nota_lengua > nota_fisica]}")
print(f"Física: {alumnos[nota_lengua < nota_fisica]}")
```

```python
# Genera un array que guarde la nota media de las tres asignaturas, redondeando a 2 decimales. Una el método ndarray.round() para redondear.
nota_media_mates = nota_mates.mean().round(2)
nota_media_lengua = nota_lengua.mean().round(2)
nota_media_fisica = nota_fisica.mean().round(2)
medias = np.array([nota_media_mates, nota_media_lengua, nota_media_fisica])
# medias = np.array({"Mates" : nota_media_mates, "Lengua": nota_media_lengua, "Física" : nota_media_fisica})
print(medias)
```

```python
# Calcula y muestra cuantos alumnos han aprobado y cuantos han suspendido en matemáticas (se aprueba con 5 o más)
num_aprobados = np.sum(nota_mates >= 5)
num_suspendidos = np.sum(nota_mates < 5)
print("Número de alumnos aprobados:", num_aprobados)
print("Número de alumnos suspendidos:", num_suspendidos)
```
